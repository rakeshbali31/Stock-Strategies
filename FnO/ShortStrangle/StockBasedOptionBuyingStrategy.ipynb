```python
from nsepython import *
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
fno_symbols = fnolist()
```

```python
fno_stocks = [symbol for symbol in fno_symbols if symbol not in ['NIFTY', 'NIFTYIT', 'BANKNIFTY']]
```

```python
def calculate_volatility(symbol):
    stock_data = yf.download(symbol + ".NS", start=(pd.to_datetime('today') - pd.DateOffset(days=365)), 
                             end=pd.to_datetime('today') + pd.DateOffset(days=1), progress=False)
    
    stock_data['Daily Returns'] = stock_data['Adj Close'].pct_change()
    
    one_year_volatility = stock_data['Daily Returns'].std() * (252 ** 0.5)
    one_month_volatility = stock_data['Daily Returns'].tail(20).std() * (20 ** 0.5)
    two_week_volatility = stock_data['Daily Returns'].tail(10).std() * (10 ** 0.5)
    
    return symbol, one_year_volatility, two_week_volatility, one_month_volatility


volatility_results = []

for symbol in fno_stocks:
    try:
        result = calculate_volatility(symbol)
        volatility_results.append(result)
    except Exception as e:
        print(f"Error processing {symbol}: {e}")

volatility_df = pd.DataFrame(volatility_results, columns=['Symbol', 'One Year Volatility', 'Two Weeks Volatility', 'One Month Volatility'])
```

```python
weekly_min_df = volatility_df.sort_values('Two Weeks Volatility', ascending=False).iloc[0:10]
```

```python
monthly_min_df = volatility_df.sort_values('One Month Volatility', ascending=False).iloc[0:10]
```
